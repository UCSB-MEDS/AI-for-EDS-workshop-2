# Ocean Temperature Time Series Project Documentation


This project visualizes and analyzes ocean temperature data collected from 2021 to 2024. The dataset includes the following columns:

- **date**: The date of each observation
- **temperature**: Measured ocean temperature in degrees Celsius
- **depth**: Depth of measurement in meters

The following sections demonstrate how to load, clean, and plot the data.

**Disclaimer**: This project contains fake data and should only be used for demonstration purposes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv("../data/ocean_temps.csv")
```

```python
data.head()
```

```python
# Plot data over time
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data.index, data['temperature'], linewidth=0.8, alpha=0.7)
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (C)')
ax.set_title('Ocean Temperature Over Time')
ax.grid(True, alpha=0.3)
    

    
plt.show()
```
